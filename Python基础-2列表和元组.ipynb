# 列表和元组


    - 概述：
        - 列表、元组等等，都是一种数据结构
        - 最基本的数据结构是序列，元组、列表都是序列的一种，还有一种是：字符串。
        - Python内置多种序列：列表、元组、字符串.........
        - 几乎所有情况下，都是可以使用列表代替元组，除了元组作为字典的键。
        - 字符串这个序列，是由字符组成的序列。
        
    - 容器：
        - 一种数据结构的基本概念：容器——可包含其他对象的对象，两种主要的容器是：序列（元组、列表），映射（字典）
        - 在序列中，每个元素都有编号（索引），在映射中每个元素都有名称（键）
        - 另外一种既不是序列，也不是映射的类型：集合（set）
        

## 通用的序列操作

#### 下面这些操作，适用于所有的序列类型，列表，元组，字符串........

    - 索引
        - 所有的索引都是从0开始，如果从末尾开始就是-1.
        - 字符串的索引可以直接操作
        
    - 切片
        - 使用切片来访问特定范围内的元素
        - 切片的索引是：包含第一个元素，不包含第二个元素
        - 前面或者后面的索引，只要是超出范围，默认包含全部
        - 切片索引顺序与索引顺序相反，切片为空 
        - 没有索引，只有冒号，默认是切片全部
        
    - 步长
        - 切片的步长默认为1，可忽略不写
        - 步长可以为负数，切片的顺序是从末尾开始
        
    - 序列相加
        - 不能拼接不同类型的序列
        
    - 乘法：
        - 是将序列重复
        
    - 成员资格检查：
        - 这是布尔运算，结果为布尔值
        - 布尔值只有两个：True   False
        - 成员资格检查，常用在检查用户输入的用户名和密码是否在已有的数据中。
        
    - 序列内置函数：
        - 常用的：len()检查序列长度，max()序列中的最大值，min()序列中的最小值。

```python
print('Hello'[2])
print('python'[4])
words = input('请输入名字')[1]
print(words)
```

```python
# 切片右边索引值超出范围
a = [1,2,3,4,5,6,7,8] 
print(a[1:11])
# 索引顺序相反，为空
print(a[4:2])
# 使用负数索引
print(a[1:-2])
#步长为负数
print(a[::-2])
```

##  主讲列表

    - 列表函数：list() 可以将一个数据生成列表。
    
    
    - 列表基本操作：
        - 给元素赋值：修改列表元素的值，直接使用列表名[索引值] = 新值
        - 给切片赋值：一次性大量修改列表元素的值
        - 删除元素：del 列表[索引值] ，被删除的元素彻底消失
        - 给切片赋值，可以在不替换原有元素的情况下，插入新元素,这种必须是切片两个索引相同才可以，插入位置在索引值后面。

```python
a = [1,5,6,7]
a[1:1] = [2,3,4] # 利用这个切片，插入了新的元素，并且原有元素保持不变
print(a)

b = [1,6,7,2,3,4,5]
b[2:2] = [9,10]
print(b)
```

### 列表方法：


    - 方法调用，需要在方法前加上句点号。
    - append()：
        - 将括号内的对象附加到列表末尾。
        - 直接在旧列表中操作，不会生成新列表。
        - 可以先创建一个空列表，然后不断的使用 append方法去添加元素。
        
    - clear():
        - 清空列表内容
        
    - copy():
        - 只是将另外一个列表名关联到这个列表地址上。
    
    - count():
        - 计算指定元素在列表中出现的次数。
        
    - extend():
        - 添加一个列表到另外一个列表的末尾。
        - 拼接是返回一个全新的列表，extend 只是在原有的列表中修改。
    - index（）：
        - 查找元素第一次出现的索引。
        
    - insert()：
        - 括号里有两个参数：插入的索引位置（原有索引的后面），插入的值。
        - 插入的值可以是列表。
        
    - pop():
        - 从列表末尾弹出一个元素
        - 最后一个元素，不管是一个数值，还是一个列表，都可以
        
    - remove():
        - 删除第一个指定值的元素（如果有多个相同值的情况下）
        
    - reverse():
        - 翻转列表排序
        - 这是一个方法，还有个函数reversed
        - reversed返回的是一个迭代器对象，需要使用list函数去生成列表
        
    - sort():
        - 对原列表进行排序，不会生成新的列表，也就是不会返回一个新的副本
        - 使用sorted函数可以生成一个新的副本。
        - sort 方法有一个隐藏参数，reverse=1 从 大到小， reverse=0 从小到大
        - 另外一个隐藏参数：key，创建一个键，这个键，一般为一个函数：len 
        - 还有隐藏的参数，那是高级用法，后期再深入学习

### 元组

    - 元组：tuple ,有一个函数 tuple（）
    - 元组的构成：圆括号，元素之间用逗号隔开
    - 空元组：一个圆括号
    - IDLE中，一个元素后面有一个逗号也可以创建一个元组。
        - 元组可以用索引值去访问元素。
        - 也可以使用切片访问。
